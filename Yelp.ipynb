# 1) Toronto

```python
# Perform API Calls - test with one restaurant 

params_to = {'term':'restaurant',
     'location':'Toronto', 
     'sort_by': 'rating',        
     'limit':'50'}
```

```python
request=requests.get(url, params=params_to, headers=headers)
print('The status code is {}'.format(request.status_code))
```

```python
#Convert the data into a JSON object
parsed_to = json.loads(request.text)
toronto = parsed_to["businesses"]
```

```python
#“businesses” is a list, enclosed by a square brackets.
# extract businesses first into a Python list and then loop over the businesses in the list.
# count = 1
name = []
phone_number = []
postal_code =[]
address = []
city = []
rating = []
category = []
cost = []
review_count = []
 

for business in toronto:
    #print("Number:", count)
    #print("Name:", business["name"])
    #print("Address:", " ".join(business["location"]["display_address"]))
    #print("City:", " ".join(business["location"]["city"]))
    #print("Rating:", business["rating"])
    #print("Category:", business["categories"][0].get('alias'))
        
    #if 'price' in business:
    #    print("Cost:", business["price"])
    #
    #print("\n")
    #print("Review Count:", business["review_count"])
    
    name.append(business["name"])
    phone_number.append(business["phone"])        
    postal_code.append(business["location"]["zip_code"])
    address.append(business["location"]["display_address"])
    city.append(business["location"]["city"])
    rating.append(business["rating"])
    category.append(business["categories"][0].get('alias'))
    
    
    if 'price' in business:
        cost.append(business["price"])
    else:
        cost.append("")
    review_count.append(business["review_count"])

    # count = count+1  
```

```python
#Display the DataFrame

df_to = pd.DataFrame(data={"Restaurant Name": name, "Restaurant Address": address, "Restaurant City": city, "Rating":
                       rating, "Category": category, "Cost": cost, "Review Count": review_count})
#df_to
df_sort_to = df_to.sort_values(by=['Rating'], ascending=False)
df_sort_to
```

```python
#Save to a CSV all data retrieved
df_sort_to.to_csv("Toronto.csv", sep=',',index=False)
#df_to.to_csv("Toronto.csv", sep=',',index=False)
```
